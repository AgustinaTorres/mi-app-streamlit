```python
#Import libraries
import os
import requests
from langdetect import detect
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound
from IPython.display import Markdown, display
```

```python
# Define functions

# Extract the video code from the URL
def get_video_code(url):
    """
    Extracts the video code from a YouTube URL.
    
    :param url: Full YouTube URL
    :return: Video code as a string
    """
    return url.split("?v=")[-1]

# Fetch the transcript from YouTube
def fetch_transcript(video_code, languages=['es', 'en']):
    """
    Tries to fetch a manually created transcript first; if not available, fetches an auto-generated one.
    
    :param video_code: YouTube video code
    :param languages: List of preferred languages in order of priority
    :return: Transcript object if found
    """
    try:
        # Attempt to retrieve manually created transcript
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_code)
        transcript = transcript_list.find_manually_created_transcript(languages)
    except (NoTranscriptFound, TranscriptsDisabled):
        try:
            # If no manual transcript, try to fetch auto-generated transcript
            transcript = transcript_list.find_generated_transcript(languages)
        except Exception as e:
            raise Exception(f"No transcripts available for this video. Error: {str(e)}")
    
    return transcript

# Transcribe the video by fetching the text from the transcript
def transcribe_youtube(transcript):
    """
    Fetches and formats the transcript into plain text.
    
    :param transcript: Transcript object
    :return: Transcribed text as a single string
    """
    try:
        transcript_data = transcript.fetch()
        return " ".join([item['text'] for item in transcript_data])
    except Exception as e:
        raise Exception(f"Error fetching transcript: {str(e)}")

# Detect the language of the given text
def detect_language(text):
    """
    Detects the language of the provided text.
    
    :param text: Text to analyze
    :return: Detected language as a string
    """
    try:
        return detect(text)
    except:
        return "Language detection failed"

# Show the transcribed text
def show_full_text(text):
    """
    Displays the transcribed text in Markdown format.
    
    :param text: Transcribed text
    """
    display(Markdown(text))

# Save the transcription to a file
def save_transcription(text, filename):
    """
    Saves the transcribed text to a file.
    
    :param text: Transcribed text
    :param filename: File path where the text will be saved
    """
    with open(os.path.join('processed_data', filename), 'w') as file:
        file.write(text)
```

```python
# Main logic to run the transcription process
if __name__ == "__main__":
    # Define the URL of the YouTube video
    url = "https://www.youtube.com/watch?v=_M-FhMORoog"
    video_code = get_video_code(url)
    
    # Attempt to fetch transcript
    try:
        transcript = fetch_transcript(video_code)
        video_text = transcribe_youtube(transcript)
        
        # Show and save the transcript
        #show_full_text(video_text)
        save_transcription(video_text, 'full_text.txt')
        
        # Detect and print the language of the transcript
        #detected_language = detect_language(video_text)
        #print(f"Detected language: {detected_language}")
    
    except Exception as e:
        print(f"Error processing video: {str(e)}")
```
